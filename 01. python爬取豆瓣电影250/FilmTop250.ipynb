```python
'''
@author:bigbear
@file:douban_movie
@time:2018/08/20 09.18
@description:
'''

import requests
import re
from bs4 import BeautifulSoup
import csv


#先创建一个csv文件，写好头部
with open("douban_top250_movies.csv",'w',encoding = 'utf-8') as filed:
    csv_writer = csv.DictWriter(filed,[
        u'电影名称',
        u'电影评分',
        u'评分人数',
        u'一句话描述',
        u'豆瓣链接',
        
    ])
    csv_writer.writeheader()
    
    
    
def get_mov_info(response):
    mov_info = {}
    soup = BeautifulSoup(response.text,'lxml')
    movies = soup.find_all('div',class_='info')
    
    for info in movies:
        mov_info['mov_name'] = info.find('span',class_='title').text
        
        mov_info['mov_link'] = info.find('a').get('href')
        
        rating_num = info.find(class_='rating_num')
        mov_info['rating_score'] = rating_num.text
        comment = rating_num.find_next_sibling().find_next_sibling()
        
        comment_num = re.findall('\d{0,}',comment.text)
        mov_info['comment_nums'] = comment_num[0]
        
        comment_one = info.find('span',class_='inq')
        if comment_one is None:
            mov_info['inq_comment'] = u' '
        else:
            mov_info['inq_comment'] = comment_one.text
        print(mov_info)
        
        write_csv(mov_info)
        
        
def write_csv(info_dict):
    with open('douban_top250_movies.csv','a+',encoding = 'utf-8') as f:
        csv_write = csv.DictWriter(f,[
            u'电影名称',
            u'电影评分',
            u'评分人数',
            u'一句话描述',
            u'豆瓣链接',
        ])
        
        csv_write.writerow({
            u'电影名称':info_dict['mov_name'],
            u'电影评分':info_dict['rating_score'],
            u'评分人数':info_dict['comment_nums'],
            u'一句话描述':info_dict['inq_comment'],
            u'豆瓣链接':info_dict['mov_link'],
        })
        
        
for num in range(0,10):
    page = num*25
    response = requests.get('https://movie.douban.com/top250?start=%d&filter='
                            % page)
    get_mov_info(response)
                
    
```
